### Even when the profiles are lousy, the resolution is not affected. 

The reason may be that the fluctuations in Krypton are much larger than the residual of the corrections. But as fluctuations decrease the effect of the correction becomes more apparent. This may be the reason why we have lousy resolution correcting at high energy
